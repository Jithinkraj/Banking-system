```python
# Banking system
class Bank:
    bankname="Bank of India"
    branch="Kuthuparamba, Kerala"

    #constructor
    def __init__(self,username,pan,mobile):
        self.username=username
        self.pan=pan
        self.mobile=mobile
        self.balance=0.0
        print(f"""
        Hi {self.username}, WELCOME to the world of next generation banking!
        """)
        
    #Account login
    def login(self,accountnumber,pin):
        self.accountnumber=accountnumber
        self.pin=pin
        if self.accountnumber==accountnumber and self.pin==pin:
            print("Login successful.")
            
        else:
            print("Invalid Account number or PIN !  Please try again.")
            
            
    #Create new Account
    def createaccount(self):
        print(f"Welcome to {Bank.bankname} , {Bank.branch}")
        username=input("Enter your name:")
        pan=input("Enter your PAN No:")
        mobile=int(input("Enter your Mobile No:"))
        address=input("Enter your address:")
        print(f"""
        
        Hello {self.username} congratulations!! Your account created successfully.
        
        """)
    
    #Deposit
    def deposit(self,amount):
        self.balance=self.balance+amount
        print(f"""
        Deposit of ₹{amount} successful. Your current balance is ₹{self.balance}
        """)

    #Withdrawal
    def withdraw(self,amount):
        if amount<self.balance:
            self.balance=self.balance-amount
            print(f"""
            Withdrawal of ₹{amount} successful. Your current balance is ₹{self.balance}.
            """)
        else:
            print("Insufficient amount! Please try again.")
        
    def bal_enquiry(self):
        print(f"""
        Your account balance is ₹{self.balance}
        
        """)
print(f"Welcome to {Bank.bankname},{Bank.branch}")
## Collecting user details to enter into banking system
username=input("Enter your name:")
pan=input("Enter your PAN number:")
mobile=int(input("Enter your Mobile number:"))

## Object creation
b=Bank(username,pan,mobile) 

## While loop used to perform options more than one time
while True:
    print(f"""
                  
                  WELCOME TO BANK OF INDIA
            {Bank.bankname} , {Bank.branch}
                 
    
    1.ACCOUNT LOGIN                    2.CREATE AN ACCOUNT
    
    
    3.AMOUNT DEPOSIT                   4.AMOUNT WITHDRAWAL  


    5.BALANCE ENQUIRY                  6.EXIT
    
    
    """)
    choice=input("Please select any option:")
    if (choice=='1'):
        accountnumber=int(input("Enter your Account Number:"))
        pin=int(input("Enter your PIN number:"))
        b.login(accountnumber,pin)
    elif(choice=='2'):
        b.createaccount()
    elif (choice=='3'):
        amount=int(input("Enter the amount to deposit:"))
        pin=int(input("Enter your PIN:"))
        b.deposit(amount)
    elif (choice=='4'):
        amount=int(input("Enter the amount to withdraw"))
        pin=int(input("Enter your PIN:"))
        b.withdraw(amount)
    elif (choice=='5'):
        b.bal_enquiry()
    elif (choice=='6'):
        print("""
        
        THANKS FOR USING BANK OF INDIA
        
        """)
        break
    else:
        print("""
        
            Invalid option! Please select a valid option
            
              """)
```
